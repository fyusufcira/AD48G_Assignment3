```r
# loading all thelibraries
library(tidyverse)
library(data.table)
```

```r
#setting seed for reproductanility
set.seed(123)

#character vector with 20 elements
char_data <- c(
  "John_Doe123", "JaneDoe@abc.com", "The_quick_brown_fox", "Lazy_Dog987",
  "123_Street_ABC", "No_symbols_here", "special$character%set", "UpperCASE123",
  "lowercase_and_numbers123", "Random_Characters!&%", "Example.with.dots123",
  "PhoneNumber_555-1234", "Email_me@example.com", "Date_12-05-2024",
  "Multiple_words_here", "WeirdMixOfNumbers123Letters", "Symbols_@_and_#",
  "someTEXTwithNUM1234", "weird.CASE.mix", "edgeCASE_here!!!"
)
```

<h1>Quantifiers

```r
#extracting all the numbers with + quantifier
numbers <- char_data %>%
  str_extract_all("\\d+")
numbers

#extracting sequences of letters and digits
optional_digit <- char_data %>%
  str_extract_all("\\w+\\d?")
optional_digit
```

<h1>Dot qualifier

```r
#replacign any character that is followed by '123' with 'ABC'
dot_example <- char_data %>%
  str_replace_all(".123", "ABC")
dot_example
```

<h1>Word Boundary (\\w)

```r
#extracting all words
word_chars <- char_data %>%
  str_extract_all("\\w+")
word_chars
```

<h1>Digit (\\d) Qualifier

```r
#checking if there are strings that contain digits
contains_digits <- char_data %>%
  str_detect("\\d")
contains_digits
```

<h1>Space (\\s) Qualifier

```r
# checking if there strings that contain spaces
contains_spaces <- char_data %>%
  str_detect("\\s")
contains_spaces
```

<h1>Character Sets and Exclusions

```r
#matching all the strings that contain any lowercase letters
lowercase_letters <- char_data %>%
  str_subset("[a-z]+")
lowercase_letters

#matching all the strings that exclude digits
no_digits <- char_data %>%
  str_subset("[^\\d]+")
no_digits
```

<h1>Anchors (^, $)

```r
#checking all the strings that start with a letter
starts_with_letter <- char_data %>%
  str_detect("^[A-Za-z]")
starts_with_letter

#checking all the strings that end with a number
ends_with_digit <- char_data %>%
  str_detect("\\d$")
ends_with_digit
```

<h1>Alternation (|)

```r
#matching strings that contain either "abc" or "123"
abc_or_123 <- char_data %>%
  str_subset("abc|123")
abc_or_123
```

<h1>Positive Lookbehind ((?<=)) and Lookahead ((?=))

```r
#extracting characters that are followed by '123' (positive lookahead)
lookahead_example <- char_data %>%
  str_extract_all(".(?=123)")
lookahead_example

#extracting characters that are preceded by a digit (positive lookbehind)
lookbehind_example <- char_data %>%
  str_extract_all("(?<=\\d).")
lookbehind_example
```
